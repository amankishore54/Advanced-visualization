```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 12,6

import seaborn as sns
```

```python
ratings = pd.read_csv('P4-Movie-Ratings.csv')
```

```python
ratings.head()
```

```python
ratings.info()
```

```python
ratings.columns = ['film', 'genre', 'criticsrating', 'audiencerating', 'budget', 'releaseyear']
```

```python
ratings.film = ratings.film.astype('category')
ratings.genre = ratings.genre.astype('category')
ratings.releaseyear = ratings.releaseyear.astype('category')
```

```python
list1= list()
mylabels = list()
for gen in ratings.genre.cat.categories:
    list1.append(ratings[ratings.genre == gen].budget)
    mylabels.append(gen)
    
sns.set_style('whitegrid') 
plt.figure(figsize=(18,9))
plt.hist(list1, bins=30, stacked=True, rwidth=1, label = mylabels)
plt.title('Distribution of budget by genre', fontsize=35, color='Black', fontname='fantasy')
plt.xlabel('Budget', fontsize=25, color='black', fontname='monospace')
plt.ylabel('Number of movies', fontsize=25, color='black', fontname='monospace')
plt.legend(fontsize=20, frameon=True, fancybox=True, shadow=True, framealpha=1)
plt.show()
```
